```python
#para graficar y analizar - fuinciona con pandas
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("netflix_titles.csv")
df
```

```python
import matplotlib.pyplot as plt
'''MUEESTRA GEAFICO Y ASU VEZ MEDIANA'''
#ahora HISTOGRAMA AGRUPANDO POR TITULO
# Histograma de la distribuciÃ³n de aÃ±os de lanzamiento
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='release_year', bins=30, kde=True)
plt.title('DistribuciÃ³n de AÃ±os de Lanzamiento')
plt.xlabel('AÃ±o de Lanzamiento')
plt.ylabel('Cantidad de TÃ­tulos')
plt.show()
```

```python
plt.figure(figsize=(8, 5)) #defino el tipo d e grafico
sns.countplot(data=df, x='type') # LE PASO DATAFRAME
plt.title('Cantidad de TÃ­tulos por Tipo')
plt.xlabel('Tipo')
plt.ylabel('Cantidad de TÃ­tulos')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='release_year')
plt.title('Cantidad de TÃ­tulos por Tipo')
plt.xlabel('Tipo')
plt.ylabel('Cantidad de TÃ­tulos')
plt.show()
```

```python
#POR RATING DE PELI
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index)
plt.title('Cantidad de TÃ­tulos por Rating')
plt.xlabel('Rating')
plt.ylabel('Cantidad de TÃ­tulos')
plt.xticks(rotation=45)
plt.show()
```

```python
# REALCION ENTRE DOS VARIABLE - SABER SI LOS AÃ‘OS MAS NUEVOS DURAN MAS QUE LOS VIEJOS
# GRAFICA DE DISPERSIÃ“N
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='release_year', y='duration', alpha=0.5)
plt.title('RelaciÃ³n entre DuraciÃ³n y AÃ±o de Lanzamiento')
plt.xlabel('AÃ±o de Lanzamiento')
plt.ylabel('DuraciÃ³n (minutos)')
plt.show()
```

```python
''''DATASET TITANIC - YA LO PROVEE LA LIB'''
titanic = sns.load_dataset("titanic")
titanic
```

```python
#EN RELACION A SEXO Y CLASE CUANTO PAGO
#sns.barplot(x="class", y="fare", hue="sex", ci=None, palette="muted", data=titanic)
sns.barplot(x="class", y="fare", hue="sex", ci=None, palette="muted", data=titanic)
```

```python

sns.histplot(data=titanic, x="age", bins=20, kde=True, color="green")
```

```python
#ANALIZAR EDAD DE PASAJEROS POR CLASE Y GENERO

plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='class', y='age', hue='sex')
plt.title('Edades de los Pasajeros por Clase y GÃ©nero')
plt.xlabel('Clase')
plt.ylabel('Edad')
plt.legend(title='GÃ©nero')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic, x='age', y='fare', hue='class', palette='viridis')
plt.title('Tarifa pagada por Edad de los Pasajeros, Coloreado por Clase')
plt.xlabel('Edad')
plt.ylabel('Tarifa')
plt.legend(title='Clase')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='class', hue='sex', palette='pastel', hue_order=['male', 'female'], dodge=False)
plt.title('Cantidad de Pasajeros Sobrevivientes por Clase y GÃ©nero')
plt.xlabel('Clase')
plt.ylabel('Cantidad de Pasajeros Sobrevivientes')
plt.legend(title='GÃ©nero')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x='age', kde=True, color='purple', bins=20)
plt.axvline(titanic['age'].mean(), color='blue', linestyle='--', label='Media')
plt.axvline(titanic['age'].median(), color='red', linestyle='-', label='Mediana')
plt.title('DistribuciÃ³n de Edades de los Pasajeros')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
#LINEA PROMEIDO ENTRE PUNTOS -> REGRESIÃ“N LINEAL

plt.figure(figsize=(10, 6))
sns.regplot(data=titanic, x='age', y='fare', scatter_kws={'alpha':0.5}, ci=95)
plt.title('RelaciÃ³n entre Edad y Tarifa Pagada')
plt.xlabel('Edad')
plt.ylabel('Tarifa')
plt.show()
```
